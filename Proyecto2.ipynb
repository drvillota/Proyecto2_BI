# Perfilamiento y preparacion de los datos

## Carga de datos

```python
#Leemos el archivo de Demografía Y Poblacion
demografia=pd.read_excel('data/Demografía Y Poblacion.xlsx')
#Leemos el archivos de Vivienda y Servicios Publicos
vivienda=pd.read_excel('data/Vivienda y Servicios Publicos.xlsx')
#Leemos el archivos de Educacion
educacion=pd.read_excel('data/Educacion.xlsx')
#Leemos el archivos de Salud
salud=pd.read_excel('data/Salud.xlsx')
```

## Perfilamiento y entendimiento de los datos

```python
#Visualizamos los atributos de los datos
demografia.info()
```

```python
demografia
```

```python
demografiaF = pd.DataFrame(demografia)
demografiaF = demografiaF.drop(demografiaF.index[0])
```

```python
#Verficamos los atributos
demografiaF['Código Departamento'] = pd.to_numeric(demografiaF['Código Departamento'], errors='coerce')
demografiaF['Código Entidad'] = pd.to_numeric(demografiaF['Código Entidad'], errors='coerce')
demografiaF['Dato Numérico'] = pd.to_numeric(demografiaF['Dato Numérico'].str.replace('.','').str.replace(',','.'), errors='coerce')
demografiaF['Dato Cualitativo'] = demografiaF['Dato Cualitativo'].apply(str)
demografiaF['Año'] = pd.to_numeric(demografiaF['Año'], errors='coerce')
demografiaF['Mes'] = pd.to_numeric(demografiaF['Mes'], errors='coerce')
```

```python
demografiaF.rename(columns = {'Código Departamento':'Código_Departamento', 'Código Entidad':'Código_Entidad', 'Dato Numérico':'Dato_Numérico', 'Dato Cualitativo':'Dato_Cualitativo', 'Unidad de Medida':'Unidad_de_Medida'}, inplace = True)
```

```python
#Borramos los datos que no aportan información 
demografiaF = demografiaF.dropna(subset=['Dato_Numérico'])
```

```python
demografiaF.info()
```

```python
demografiaF.describe()
```

```python
#Guardamos el dataset limpio
demografiaF.to_csv('data/Demografía Y Poblacion Clean.csv' , encoding = 'latin-1', sep=',', index=False)
```
